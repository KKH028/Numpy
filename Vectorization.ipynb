```python
import numpy as np

a = np.array([1,2,3,4])
print(a)
```

```python
import time

a = np.random.rand(10000000)
b = np.random.rand(10000000)


# print(type(a))
# print(a.shape)
# print(a.ndim)
# print(type(b))
# print(b.shape)

tic = time.time()
c = np.dot(a, b)  # dot product function support vertorization
toc = time.time()
print("tic=", tic)
print("toc=", toc)
print(c)
print("vectorized version:" + str(1000*(toc-tic)) + "ms")
```

```python
c = 0
tic = time.time() #시스템 부팅 이후 현재까지의 경과 시간을 us로 표시
for i in range(1000000):
  c += a[i] * b[i] # dot product
toc = time.time()
print(c)
print("non vectorized version:" + str(1000*(toc-tic)) + "ms")
```
